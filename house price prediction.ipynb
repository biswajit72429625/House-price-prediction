```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['month']=df['date'].dt.month
```

```python
df['year']=df['date'].dt.year
```

```python
df.corr()['price'].sort_values()
```

```python
df.drop(['date','month'],inplace=True,axis=1)
```

```python
df.info()
```

```python
x=df.drop('price',axis=1).values
y=df.price.values
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
mod=Sequential()
```

```python


mod.add(Dense(19,activation='relu'))
mod.add(Dense(19,activation='relu'))
mod.add(Dense(19,activation='relu'))
mod.add(Dense(19,activation='relu'))
mod.add(Dense(1))

mod.compile(optimizer='adam',loss='mse')
```

```python
mod.fit(xtrain,ytrain,epochs=300,validation_data=(xtest,ytest))
```

```python
plt.plot(pd.DataFrame(mod.history.history))
```

```python
pred=mod.predict(xtest)
```

```python
sns.scatterplot(pred.reshape(5400,),ytest)
```

```python
from tensorflow.keras.models import load_model
```

```python
mod.save('kc_housing.h5')
```

```python
load=load_model('kc_housing.h5')
```

```python
comp=load.predict(xtest)
```

```python
comp=comp.reshape(len(comp),)
```

```python
from sklearn.metrics import mean_squared_error
```
